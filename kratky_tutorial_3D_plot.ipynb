# Ako plotiť v 3D - tutoriál pre Ondru
Majme 2 matice, ktoré sú 2D. Fakticky sú to obrázky s pixelmi, ako z fotoaparátu. Jedna matica obsahuje hodnoty reciprokého priestoru, nazv. q_matica. Druhá matica nech obsahuje pixely aberačnej funkcie.

## Predspracovanie dát
Treba vedieť, ako 3D plotenie hodnôt v Pythone funguje, aby sme boli schopní si niečo z tohto ukázať v 3D. Preto sem hodím nejaký príklad a Ty potom máš za úlohu to skúsiť nakódiť u seba. :)

```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# majme rozlisenie medzi bodmi v reciprockom priestore
q_point_to_point_resolution = 1.245786785 # - dokonca ani nepotrebujeme ručne plniť maticu q bodov, ak poznáme point to point resolution v q. viď nižšie 
# majme pole (vektor) hodnot q v 2 smeroch
q_1 = np.arange(-5, 5, q_point_to_point_resolution)
q_2 = np.arange(-5, 5, q_point_to_point_resolution)
# meshgrid - mriežka. Predstav si naozajstnú mriežku alebo sieť s uzlami. Takú vytvoríme príkazom np.meshgrid() - transformujeme q_1 a q_2 na mriežky
# pre viac info si vyploť q_1 a q_2 po meshgride a vygoogli meshgrid
q_1, q_2 = np.meshgrid(q_1, q_2)

# náhodná matica
image = np.zeros(shape= (np.size(q_1, 0), np.size(q_2, 0) ))
for i in range(np.size(image, 0)):
    for j in range(np.size(image, 1)):
        image[i, j] = j**2

# 3D plot
# surface - povrch, wireframe - taká sieťka v 3D, countour - kontúra. môžeš zakomentovať rôzne štýly a skúsiť ich
surf = ax.plot_surface(q_1, q_2, image)
#surf = ax.plot_wireframe(q_1, q_2, image)
#surf = ax.contour3D(q_1, q_2, image)


# Zmena limitov
ax.set_zlim(0, 90)

# color bar
fig.colorbar(surf)

plt.show()
```
